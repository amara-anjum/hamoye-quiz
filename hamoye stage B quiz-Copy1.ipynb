```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('energydata_complete.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['date','lights'],axis=1)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
normalised_df
```

```python
x = normalised_df.drop(columns=['Appliances'],axis=1)
y = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
test_size=0.3, random_state=1)
x_train.shape
```

```python
reg_df=normalised_df[['T2','T6']]
reg_df.head()
```

```python
x1=reg_df['T2'].values.reshape(-1,1)
y1=reg_df['T6'].values.reshape(-1,1)
x1_train, x1_test, y1_train , y1_test= train_test_split(x1,y1, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

linear_model.fit(x1_train, y1_train)

y_pred1 = linear_model.predict(x1_test)
y_pred1
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y1_test, y_pred1)
round(r2, 2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x_train, y_train)

y_pred = reg.predict(x_test)
```

```python
train_score=reg.score(x_train , y_train)
test_score= reg.score(x_test , y_test)
print('training score:{:.3f}'.format(train_score))
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
round(mae, 2)
print('MAE: {}'.format(round(mae,2)))
```

```python

rss = np.sum(np.square(y_test - y_pred))
print('residual sum of squares:{}'.format(round(rss,2)))
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
round(rmse, 3) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
round(r2_score, 3)
```

```python
def weights(model, feat,col_name):
    weights= pd.Series(model.coef_ ,feat.columns).sort_values()
    weights_df=pd.DataFrame(weights).reset_index()
    weights_df.columns=['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df 
```

```python
linear_reg_weights= weights(reg, x_train, 'Linear_Model_Weight' )
linear_reg_weights
```
